```python
def get_artist(name):
    results = sp.search(q='artist:' + name, type='artist')
    items = results['artists']['items']
    if len(items) > 0:
        return items[0]
    else:
        return None
```

```python
g_types = []
for keys in d:
    g_types.append(d[keys])
g_types
```

```python
import nltk
from nltk import FreqDist
#find the most common genres
pop_genres = FreqDist(list(chain.from_iterable(g_types)))
```

Saving the frequency count for ech of the genres by artists in order to narrow the genre selection to the most commonly used ones

```python
for key in d:
    new_d = {}
    if len(d[key]) > 1:
        for k in d[key]:
            new_d[k] = pop_genres[k]
        d[key] = new_d
    else:
        b = d[key][0]
        new_d[b] = pop_genres[b]
        d[key] = new_d
```

Picking the most popular genres among the list by artists

```python
import operator
artist_genres = {}
for keys in d:
    standard_genre = max(d[keys].items(), key=operator.itemgetter(1))[0]
    artist_genres[keys] = standard_genre
artist_genres
```

```python
artist_genres2 = {k.lower(): v for k,v in artist_genres.items()}
```

```python
#artist_genres2 = {k.lower(): v for k,v in artist_genres.items()}
#artist_genres3 = {keys.replace('the ',''):values for keys,values in artist_genres2.items() if in unfound}
#artist_genres3 = {keys.replace('the ',''):values in unfound for keys,values in artist_genres2.items()}
artist_genres3 = {keys.replace('the ',''):values for keys,values in artist_genres2.items()}
```

```python
#replace 'beatles' to 'the beatles'
artist_genres3['the beatles'] = artist_genres3.pop('beatles')
```

```python
artist_genres3['glee'] = artist_genres3.pop('glee cast')
artist_genres3['the killers'] = artist_genres3.pop('killers')
```
